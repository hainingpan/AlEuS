```python
import numpy as np
import kwant
import scipy.sparse.linalg as sla
from scipy.constants import hbar, m_e, e, eV
```

```python
sigma_0 = np.array([[1, 0], [0, 1]]) 
sigma_x = np.array([[0, 1], [1, 0]])
sigma_y = np.array([[0, -1j], [1j, 0]])
sigma_z = np.array([[1, 0], [0, -1]])

tau_0 = np.array([[1, 0], [0, 1]]) 
tau_x = np.array([[0, 1], [1, 0]])
tau_y = np.array([[0, -1j], [1j, 0]])
tau_z = np.array([[1, 0], [0, -1]])
```

```python
def t_function(m_eff=1, a=0.1):
    """
    Parameters:
    ------------
    m_eff: the effective mass of the conduction band, in unit of m_e
    a: lattice constant, in unit of nm
    
    Returns:
    ----------
    t: hopping amplitude in the tight-binding model, in unit of meV
    """
    
    t = hbar**2 / (2 * m_eff * m_e * (a * 1e-9)**2 * 1e-3 * eV) # in unit of meV
    return t

def alpha_function(alpha_eVA=0.5, a=10):
    """
    Parameters:
    ------------
    alpha_eVA: strength of spin-orbit coupling, in unit of eVA
    a: lattice constant, in unit of nm
    
    Return:
    ---------
    alpha: spin-orbit hopping amplitude in the tight-binding model, in unit of meV
    """
    
    alpha = alpha_eVA * 1e2 / (2 * a)
    return alpha

def get_V(positions, potential):
    """
    Parameters:
    --------------
    positions: N by 2 ndarray of (x, y) coordiantes of lattice points
    potential: (N, ) ndarray of the onsite potential, in unit of Volts
    
    Returns:
    ----------
    V: a closure function, which takes (x, y) and returns the potential in unit of meV on that site
    """
    def V(x, y):
        r = np.array([[x, y]])
        sqdist = np.sum((positions - r)**2, axis=1) # find the point closet to (x, y)
        return -1e3 * potential[np.argmin(sqdist)]
    return V

def get_positions(syst):
    number_of_sites = len(syst.sites)
    dimension = len(syst.sites[0].pos)
    positions = np.zeros((number_of_sites, dimension))
    for i, site in enumerate(syst.sites):
        positions[i, :] = site.pos
    return positions

def local_dos(wf, norbs=4):
    wfsq = np.abs(wf)**2
    ldos = np.sum(wfsq.reshape(-1, norbs), axis=1)
    return ldos

def get_Gaussian_fun(E0, Gaussian_width):
    def Gaussian_fun(E):
        return (np.sqrt(2*np.pi) * Gaussian_width)**-1 * np.exp(-(E - E0)**2 / (2 * Gaussian_width**2))
    return Gaussian_fun
```

```python
def make_bare_Al(a=0.1, thickness_Al=6, width=50):
    '''
    Parameters:
    -------------
    a: lattice constant of Al in unit of nm
    thickness_Al: thickness of Al in unit of nm
    width: width of Al in unit of nm
    
    Returns:
    -------------
    the finalized kwant syst
    '''
    
    def shape(pos):
        x, y = pos
        in_x = 0 <= x <= width
        in_y = 0 <= y <= thickness_Al
        return in_x and in_y
    
    def onsite(site, t_Al, E_F, V_Al, Delta, Ez, k, is_BdG):
        x, y = site.pos
        if is_BdG:
            return (6 * t_Al - E_F + V_Al(x, y) - 2 * t_Al * np.cos(k)) * np.kron(tau_z, sigma_0) \
                + Ez * np.kron(tau_0, sigma_z) + Delta * np.kron(tau_x, sigma_0)
        else:
            return (6 * t_Al - E_F + V_Al(x, y) - 2 * t_Al * np.cos(k)) * sigma_0 + Ez * sigma_z
         
    def hopping(site1, site2, t_Al, is_BdG):
        x1, y1 = site1.pos
        x2, y2 = site2.pos
        if is_BdG:
            return -t_Al * np.kron(tau_z, sigma_0)
        else:
            return -t_Al * sigma_0
        
    lat = kwant.lattice.square(a)
    syst = kwant.Builder()
    syst[lat.shape(shape, (0, 0))] = onsite
    syst[lat.neighbors()] = hopping
    return syst.finalized()
```

```python
def make_Al_EuS(a=0.1, thickness_Al=6, thickness_EuS=2, width=60):
    '''
    Parameters:
    -------------
    a: lattice constant in unit of nm, same for Al and EuS
    thickness_Al: thickness of Al in unit of nm
    thickness_EuS: thickness of EuS in unit of nm
    width: width of Al and EuS in unit of nm
    
    Returns:
    ----------
    finalized kwant syst
    '''
    
    def shape(pos):
        x, y = pos
        in_x = 0 <= x <= width 
        in_y = -thickness_EuS <= y <= thickness_Al
        return in_x and in_y
    
    def onsite(site, t_Al, E_F, V_Al, Delta, t_EuS, E_gap, h, t_tunnel, k, is_BdG):
        x, y = site.pos
        if is_BdG:
            pauli_matrix_kinetic = np.kron(tau_z, sigma_0)
            pauli_matrix_zeeman = np.kron(tau_0, sigma_z)
            pauli_matrix_sc = np.kron(tau_x, sigma_0)
        else:
            pauli_matrix_kinetic = sigma_0
            pauli_matrix_zeeman = sigma_z
            pauli_matrix_sc = 0.0 * sigma_0
        
        if y > a:
            return (6 * t_Al - E_F + V_Al(x, y) - 2 * t_Al * np.cos(k)) * pauli_matrix_kinetic \
                + Delta * pauli_matrix_sc
        elif y == a:
            return (5 * t_Al + t_tunnel - E_F + V_Al(x, y) - 2 * t_Al * np.cos(k)) * pauli_matrix_kinetic \
                + Delta * pauli_matrix_sc
        elif y == 0:
            return (5 * t_EuS + t_tunnel + E_gap - 2 * t_EuS * np.cos(k)) * pauli_matrix_kinetic \
                + h * pauli_matrix_zeeman
        else:
            return (6 * t_EuS + E_gap - 2 * t_EuS * np.cos(k)) * pauli_matrix_kinetic \
                + h * pauli_matrix_zeeman
        
    def hopping(site1, site2, t_Al, t_EuS, t_tunnel, is_BdG):
        x1, y1 = site1.pos
        x2, y2 = site2.pos
        if is_BdG:
            pauli_matrix_kinetic = np.kron(tau_z, sigma_0)
        else:
            pauli_matrix_kinetic = sigma_0
        
        if y1 >= a and y2 >= a:
            return -t_Al * pauli_matrix_kinetic
        elif (y1 == a and y2 == 0) or (y1 == 0 and y2 == a):
            return -t_tunnel * pauli_matrix_kinetic
        else:
            return -t_EuS * pauli_matrix_kinetic
        
    lat = kwant.lattice.square(a)
    syst = kwant.Builder()
    syst[lat.shape(shape, (0, 0))] = onsite
    syst[lat.neighbors()] = hopping
    return syst.finalized()
```

```python
Al_EuS_geo_params = dict(
    a=0.1,
    thickness_Al=6,
    thickness_EuS=2,
    width=50
)
Al_EuS = make_Al_EuS(**Al_EuS_geo_params)
```

```python
positions_hybrid = get_positions(syst=Al_EuS)
SC_indices = []
FM_indices = []
for j in range(len(positions_hybrid[:, 1])):
    if positions_hybrid[j, 1] > 0:
        SC_indices.append(j)
    else:
        FM_indices.append(j)
```

```python
bare_Al_geo_params = dict(
    a=Al_EuS_geo_params['a'],
    thickness_Al=Al_EuS_geo_params['thickness_Al'],
    width=Al_EuS_geo_params['width']
)
bare_Al = make_bare_Al(**bare_Al_geo_params)

np.random.seed(0)
U_dis = 1
positions_Al = get_positions(syst=bare_Al)
num_sites_Al = positions_Al.shape[0]
potential_Al = 2 * U_dis * (np.random.rand(num_sites_Al) - 0.5)
V_Al = get_V(positions=positions_Al, potential=potential_Al)
```

```python
plt.figure(figsize=(10, 1.2))
plt.scatter(positions_Al[:, 0], positions_Al[:, 1], c=potential_Al, s=0.2, vmax=1, vmin=-1, cmap='coolwarm')
plt.colorbar(ticks=[-1, 0, 1])
plt.show()
```

```python
t_Al = t_function(m_eff=1, a=Al_EuS_geo_params['a'])
t_EuS = t_function(m_eff=0.3, a=Al_EuS_geo_params['a'])
t_tunnel = t_function(m_eff=np.sqrt(0.3 * 1), a=Al_EuS_geo_params['a'])

Al_EuS_ham_params = dict(
    t_Al=t_Al,
    E_F=11e3,
    V_Al=V_Al,
    Delta=0.35,
    t_EuS=t_EuS,
    E_gap=0.5e3,
    h=30,
    t_tunnel=t_tunnel,
    k=0.0,
    is_BdG=True
)

N_energies = 60
```

```python
H_bdg = Al_EuS.hamiltonian_submatrix(params=Al_EuS_ham_params, sparse=True)
es, wfs = sla.eigsh(A=H_bdg, k=N_energies, which='LM', sigma=0.0, return_eigenvectors=True)
```
